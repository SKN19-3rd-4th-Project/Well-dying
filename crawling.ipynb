```python
import requests
from bs4 import BeautifulSoup

url = "https://www.kakaocorp.com/page/detail/10876"
headers = {"User-Agent": "Mozilla/5.0"}

resp = requests.get(url, headers=headers)
resp.raise_for_status()
html = resp.text
soup = BeautifulSoup(html, "html.parser")
```

```python
content_div = soup.find("div", class_="wrap_cont")
```

```python
paragraphs = [p.get_text(strip=True) for p in content_div.find_all("p")]

text = "\n".join(paragraphs)

print(text)
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

url = "https://help.naver.com/service/5640/contents/17441"

options = webdriver.ChromeOptions()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

driver.get(url)

time.sleep(2)  # JS 렌더링 대기

soup = BeautifulSoup(driver.page_source, "html.parser")

content_div = soup.find("div", class_=lambda x: x and x.startswith("HelpListItem_detail_box"))
text = content_div.get_text("\n", strip=True)

print(text)
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# 1. 크롬 옵션 설정 (화면 안 띄우고 돌리고 싶으면 headless 사용)
options = Options()
options.add_argument("--headless=new")

driver = webdriver.Chrome(options=options)

try:
    # 2. 네이버 도움말 새 URL
    url = "https://help.naver.com/service/5640/contents/9545?lang=ko&osType=COMMONOS"
    driver.get(url)

    # 3. JS 로딩 시간 조금 여유주기
    time.sleep(2)

    # 4. 본문이 들어있는 div 선택
    #    - class 이름 뒤가 바뀔 수 있어서 앞부분만 공통으로 잡음
    elem = driver.find_element(By.CSS_SELECTOR, "div[class^='HelpListItem_detail_box']")

    # 5. div 안의 전체 텍스트 추출
    text = elem.text
    print(text)

finally:
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# 1. 크롬 옵션 설정 (화면 안 띄우고 돌리고 싶으면 headless 사용)
options = Options()
options.add_argument("--headless=new")

driver = webdriver.Chrome(options=options)

try:
    # 2. URL
    url = "https://tipoff-thanos.kr/entry/%EC%82%AC%EB%A7%9D-%ED%9B%84-%EC%98%A8%EB%9D%BC%EC%9D%B8-%EC%87%BC%ED%95%91-%EA%B3%84%EC%A0%95%EA%B3%BC-%ED%8F%AC%EC%9D%B8%ED%8A%B8%EB%8A%94-%EC%96%B4%EB%96%BB%EA%B2%8C-%EB%90%A0%EA%B9%8C-1?utm_source=chatgpt.com"
    driver.get(url)

    # 3. JS 로딩 시간 조금 여유주기
    time.sleep(2)

    # 4. 본문이 들어있는 div 선택
    #    - class 이름 뒤가 바뀔 수 있어서 앞부분만 공통으로 잡음
    elem = driver.find_element(
        By.CSS_SELECTOR,
        "div.tt_article_useless_p_margin.contents_style"
    )

    # 5. div 안의 전체 텍스트 추출
    text = elem.text
    print(text)

finally:
    driver.quit()
```
