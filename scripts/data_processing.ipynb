```python
# pdfplumber

import pdfplumber
gong0 = '../data/raw/전국_공영장례_조례집(서울_인천_경기).pdf'

with pdfplumber.open(gong0) as pdf:
    texts = []
    for page in pdf.pages:
        try:
            texts.append(page.extract_text())
        except:
            texts.append("")  
    pdfplumber_text = '\n'.join(texts)
```

```python
import os
CLOVA_API_URL = os.getenv("CLOVA_API_URL");
CLOVA_SECRET_KEY = os.getenv("CLOVA_SECRET_KEY");       # 쥬피터 상에서 출력 방지로 세미콜론 붙임 사실 필요없음
```

```python
from pdf2image import convert_from_path
import requests
import uuid
import json
import time
from PIL import Image
import io

api_url = CLOVA_API_URL
secret_key = CLOVA_SECRET_KEY

def pdf_to_text_with_ocr(pdf_path, api_url, secret_key):
    """
    PDF → 이미지 변환 → OCR → 텍스트
    
    Args:
        pdf_path: PDF 파일 경로
        api_url: 클로바 OCR API Gateway URL
        secret_key: 클로바 OCR Secret Key
    
    Returns:
        str: 추출된 전체 텍스트 (페이지별로 \n\n 구분)
    """
    
    # 1. PDF를 이미지로 변환
    images = convert_from_path(pdf_path, dpi=300)  # 해상도 높게
    
    # 전체 페이지의 텍스트를 저장할 리스트    
    all_text = []
    
    # 2. 각 페이지마다 OCR
    for i, img in enumerate(images):
        print(f"Processing page {i+1}...")
        
        # PIL Image 객체를 BytesIO (메모리 상의 바이너리)로 변환
        img_byte_arr = io.BytesIO()
        img.save(img_byte_arr, format='PNG')    # PNG 형식으로 저장
        img_byte_arr.seek(0)    # 파일 포인터를 처음으로 되돌림
        
        # OCR API 요청 데이터 구성
        request_json = {
            'images': [{'format': 'png', 'name': f'page_{i+1}'}],   # 이미지 정보
            'requestId': str(uuid.uuid4()), # 고유 요청 ID (랜덤 생성)
            'version': 'V2',    # API 버전
            'timestamp': int(time.time() * 1000)    # 현재 시간 (밀리초)
        }
        
        # 요청 본문 (message)과 파일 데이터 준비
        payload = {'message': json.dumps(request_json).encode('UTF-8')}
        files = [('file', img_byte_arr)]    # 이미지 바이너리
        headers = {'X-OCR-SECRET': secret_key}  # 인증 헤더
        
        # OCR API 호출 (POST 요청)
        response = requests.post(api_url, headers=headers, data=payload, files=files)
        
        # 응답 성공 시 텍스트 추출
        if response.status_code == 200:
            result = response.json()    # JSON 응답 파싱
            # 모든 필드의 inferText를 공백으로 연결
            page_text = ' '.join([field['inferText'] 
                                  for field in result['images'][0]['fields']])
            all_text.append(page_text)  # 해당 페이지 텍스트 저장
        else:
            # 실패 시 에러 출력
            print(f"Page {i+1} failed: {response.status_code}")
            
    # 모든 페이지 텍스트를 \n\n로 구분하여 반환
    return '\n\n'.join(all_text)
```

```python
def pdf_to_text_with_table_ocr(pdf_path, api_url, secret_key):
    """
    PDF → 이미지 변환 → 표 인식 OCR → 텍스트
    """
    # 1. PDF를 이미지로 변환
    images = convert_from_path(pdf_path, dpi=300)
    
    all_text = []
    
    # 2. 각 페이지마다 OCR
    for i, img in enumerate(images):
        print(f"Processing page {i+1}...")
        
        img_byte_arr = io.BytesIO()
        img.save(img_byte_arr, format='PNG')
        img_byte_arr.seek(0)
        
        # ✅ 수정 1: enableTableDetection 추가!
        request_json = {
            'images': [{'format': 'png', 'name': f'page_{i+1}'}],
            'requestId': str(uuid.uuid4()),
            'version': 'V2',
            'timestamp': int(time.time() * 1000),
            'enableTableDetection': True  # 표 인식 활성화!
        }
        
        payload = {'message': json.dumps(request_json).encode('UTF-8')}
        files = [('file', img_byte_arr)]
        headers = {'X-OCR-SECRET': secret_key}
        
        response = requests.post(api_url, headers=headers, data=payload, files=files)
        
        # ✅ 수정 2: 표 감지 여부에 따라 다르게 처리!
        if response.status_code == 200:
            result = response.json()
            
            # 표가 감지되었는지 확인
            if 'tables' in result['images'][0] and result['images'][0]['tables']:
                # 표 데이터 추출
                page_text_parts = []
                for table in result['images'][0]['tables']:
                    for cell in table['cells']:
                        cell_text = ' '.join([
                            word['inferText'] 
                            for line in cell['cellTextLines']
                            for word in line['cellWords']
                        ])
                        if cell_text.strip():
                            page_text_parts.append(cell_text)
                page_text = '\n'.join(page_text_parts)
            else:
                # 표 없으면 일반 텍스트
                page_text = ' '.join([field['inferText'] 
                                      for field in result['images'][0]['fields']])
            
            all_text.append(page_text)
        else:
            print(f"Page {i+1} failed: {response.status_code}")
    
    return '\n\n'.join(all_text)
```

```python
def clean_ocr_text(text):
    """
    OCR 결과에서 불필요한 부분 제거하고 정리
    """
    import re
    
    # 지원내용 섹션만 추출
    pattern = r'지원내용.*?(?=장사시설|$)'
    match = re.search(pattern, text, re.DOTALL)
    
    if match:
        support_section = match.group()
        
        # 그림 설명 부분 제거 (NEWS, 영안실 안치료 및 운구비 등이 섞인 부분)
        support_section = re.sub(r'NEWS.*?(?=2\.)', '', support_section, flags=re.DOTALL)
        
        return support_section
    
    return text
```

```python
import fitz  # PyMuPDF

def extract_text_by_area(pdf_path):
    """
    영역별로 텍스트 추출 (드래그 복사처럼)
    """
    doc = fitz.open(pdf_path)
    page = doc[18]
    
    # 페이지를 영역별로 나눠서 추출
    width = page.rect.width
    height = page.rect.height
    
    # 4등분 영역 정의
    areas = [
        fitz.Rect(0, 0, width/4, height),              # 왼쪽 1
        fitz.Rect(width/4, 0, width/2, height),        # 왼쪽 2
        fitz.Rect(width/2, 0, 3*width/4, height),      # 오른쪽 1
        fitz.Rect(3*width/4, 0, width, height)         # 오른쪽 2
    ]
    
    texts = []
    for i, area in enumerate(areas):
        text = page.get_text("text", clip=area)  # 특정 영역만 추출
        texts.append(f"{i+1}. {text.strip()}")
    
    return '\n'.join(texts)
```

```python

file_configs = [{
        "dataname": public_funeral_ordinance,
        "name": "public_funeral_ordinance",
        "type": "Public_Funeral_Ordinance", # 공영 장례 조례
},
{
        "dataname": cremation_words_support,
        "name": "cremation_words_support",
        "type": "Cremation_Subsidy_Ordinance", # 화장 장려금 단어 사전
        "sub_type" : "cremation_words_support"
},
{
        "dataname": cremation_detail,
        "name": "cremation_detail",
        "type": "Cremation_Subsidy_Ordinance", # 화장 장려금 조례 내용
        "sub_type" : "cremation_detail"
},
{
        "dataname": cremation_etcetera,
        "name": "cremation_etcetera",
        "type": "Cremation_Subsidy_Ordinance", # 화장 장려금 기타 지역
        "sub_type" : "cremation_etcetera"
},
{
        "dataname": cremation_search,
        "name": "cremation_search",
        "type": "Cremation_Subsidy_Ordinance", # 화장 장려금 조례 검색 방법
        "sub_type" : "cremation_search"
}]

# VectorStore 초기화
vectorstore = PineconeVectorStore(
    index_name=INDEX_NAME,
    embedding=embeddings
)
region_by_dataname = {}
# 데이터 업로드
for config in file_configs:
    texts = config["dataname"]
    regions = set()
    # 네임스페이스 결정
    if config["name"] in ["cremation_search", "cremation_words_support"]:
        namespace = "guide"
    else:
        namespace = "ordinance"
    # 메타데이터 생성
    metadatas = []
    for i, text in enumerate(texts):
        meta = {"type": config["type"]}
        if "sub_type" in config:
            meta["sub_type"] = config["sub_type"]

        if texts == public_funeral_ordinance:
            lines = text.split('\n')
            if len(lines) >= 2:
                region = lines[1].strip()
                if "공영장례" in region:
                    region = region[:region.index("공영장례")]
                if "저소득층" in region:
                    region = region[:region.index("저소득층")]
                if "무연고" in region:
                    region = region[:region.index("무연고")]
                meta['region'] = region.strip()
                regions.add(region.strip())

        if texts == cremation_detail:
            lines = text.split('\n')
            if len(lines) >= 2:
                if i % 2 == 0:
                    region = lines[1]
                    region = region.replace(" ｜ ", " ")
                if "화장" in region:
                    region = region[:region.index('화장')]
                if "장사" in region:
                    region = region[:region.index('장사')]
                if "장묘" in region:
                    region = region[:region.index('장묘')]
                if "군민" in region:
                    region = region[:region.index('군민')]
                meta['region'] = region.strip()
                regions.add(region.strip())

        if texts == cremation_etcetera:
            lines = text.split('\n')
            if len(lines) >= 0:
                region = lines[0]
                if "장사" in region:
                    region = region[:region.index('장사')]
                if "남도광역추모공원" in region:
                    region = region[:region.index('남도광역추모공원')]
                if "종합" in region:
                    region = region[:region.index('종합')]
                if "공설" in region:
                    region = region[:region.index('공설')]
                region = region[1:]                
                meta['region'] = region.strip()
                regions.add(region.strip())  

        metadatas.append(meta)

    # 자동 임베딩 + 업로드
    vectorstore.add_texts(
        texts=texts,
        metadatas=metadatas,
        namespace = namespace
    )
    dataname_key = config["name"]
    region_by_dataname[dataname_key] = sorted(list(regions))

    print(f"{config['type']} 업로드 완료 ({len(texts)}건)")

print("전체 업로드 완료")
```
